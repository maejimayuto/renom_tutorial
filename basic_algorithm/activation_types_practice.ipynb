```python
import renom as rm
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=[x/100 for x in range(-3000,3000)]
input1=np.array(data)

#for confirmation
print(input1[0:5])
```

```python
def plt_act(x,ymin=-3,ymax=3):
    plt.grid()
    plt.plot(input1,x)
    plt.xlim(-30,30)
    plt.ylim(ymin,ymax)
    plt.xlabel('input')
    plt.ylabel('output')
    plt.show()
```

```python
output3=rm.relu(input1) # Sequential ->rm.Relu()
plt_act(output3)
```

```python
output4=rm.leaky_relu(input1) # Sequential ->rm.Leaky_Relu()
plt_act(output4)
```

```python
output5=rm.leaky_relu(input1,0.5)
plt_act(output5)
```

```python
output6=rm.elu(input1) # Sequential ->rm.Elu()
plt_act(output6)
```

```python
output7=rm.elu(input1,0.5)
plt_act(output7)
```

```python
output8=rm.selu(input1) # Sequential ->rm.Selu()
plt_act(output8)
```
