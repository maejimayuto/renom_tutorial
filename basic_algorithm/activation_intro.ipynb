```python
import renom as rm
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.array([i/100 for i in range(-1000,1000)])
y=rm.sigmoid(x) # Sequential ->rm.Sigmoid()
plt.grid()
plt.plot(x,y)
plt.xlabel('input')
plt.ylabel('output')
plt.show()
```

```python
y=rm.tanh(x) # Sequential ->rm.Tanh()
plt.grid()
plt.plot(x,y)
plt.xlabel('input')
plt.ylabel('output')
plt.show()
```

```python
y=rm.relu(x) # Sequential ->rm.Relu()
plt.grid()
plt.plot(x,y)
plt.xlabel('input')
plt.ylabel('output')
plt.show()
```
